#Pairplot is made using seaborn library

```python
sns.pairplot(df)
```

#A heat map is plotted

```python

corrmat = df.corr()
hm = sns.heatmap(corrmat, 
                 cbar=True, 
                 annot=True, 
                 square=True, 
                 fmt='.1f', 
                 annot_kws={'size': 5}, 
                 yticklabels=df.columns, 
                 xticklabels=df.columns, 
                 )
plt.show()
```
